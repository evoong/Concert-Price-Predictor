```python
import pandas as pd
```

```python
path = "../../Documents/Ticket Sales.xlsx"

stubhub = pd.read_excel(path, sheet_name ="stubhub")
ticketweb = pd.read_excel(path, sheet_name ="TicketWeb")
```

```python
market = stubhub.join(ticketweb[["Artist", "Tier", "TicketWeb"]].set_index('Artist'), on='Artist', how='outer') 
market.iloc[ :, 1:7].sort_values(['Dates'], ascending=True)
```

```python
market["profit"] = market["Stubhub_1"] - market["TicketWeb"]
market[ (market['TicketWeb'].isnull()== False)].sort_values(['Dates'], ascending=True)
```

```python
# Filter the market DataFrame to show where Stubhub is cheaper than Me
cheaper_stubhub = market[market['Stubhub_1'] < market['Me']]


print(cheaper_stubhub)
```
